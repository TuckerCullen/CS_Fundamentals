# Dijkstra's Shortest Path Algorithm 

Dijkstra's shortest path algorithm find the shortest path from a specificed start vertex to every other vertex in the graph. The algorithm works as long as the graph is fully connected and has non-negative weights (it doesn't matter if it's directed or undirected).
